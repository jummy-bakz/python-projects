```python
# Top 10 active cases in Africa country
```

```python
# Top 10 active cases in Africa country
Top_10_active_cases = Africa.groupby(by = 'countriesAndTerritories').max()[['Active cases','year']].sort_values(by=['Active cases'] ,ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title('Top 10 Africa cases',size=25)
ax=sns.barplot(data=Top_10_active_cases.iloc[:10],y='Active cases',x='countriesAndTerritories',linewidth=2,edgecolor='red')
plt.xlabel('countriesAndTerritories')
plt.ylabel('Active cases')
plt.show()
```

```python
# Top 10 Africa country with highest deaths
```

```python
Top_10_deaths = Africa.groupby(by = 'countriesAndTerritories').max()[['deaths','year']].sort_values(by=['deaths'] ,ascending=False).reset_index()
fig=plt.figure(figsize=(18,5))
plt.title('Top 10 deaths in Africa',size=25)
ax=sns.barplot(data=Top_10_deaths.iloc[:10],y='deaths',x='countriesAndTerritories',linewidth=2,edgecolor='black')
plt.xlabel('countriesAndTerritories')
plt.ylabel('deaths')
plt.show()
```
